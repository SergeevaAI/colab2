2.3.3. Напишите программирование на Python для создания круговой диаграммы с названием популярности языков программирования. Сделайте несколько клиньев круговой диаграммы.  
Пример данных:
Языки программирования: Java, Python, PHP, JavaScript, C #, C ++
Популярность: 22.2, 17.6, 8.8, 8, 7.7, 6.7
Фрагмент кода дает вывод, показанный на следующем снимке экрана:

image.png

```python
from matplotlib import pyplot as plt
import numpy as np
languages ='Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'
popuratity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
fig=plt.figure(figsize=(8, 6))
ax=fig.add_axes([0, 0, 1, 1])
colors = ["dodgerblue", "darkorange", "green", "r", "mediumorchid", "sienna"]
explode = (0.1, 0, 0, 0, 0, 0.1)  
plt.pie(popuratity, explode=explode, labels=languages, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140, wedgeprops={'edgecolor':"k"})
ax.set_title("PopularitY of Programming Language\n" + "Worldwide, Oct 2017 compared to a year ago", pad=5, bbox=dict(facecolor='silver', edgecolor='black'))
plt.show()
```

2.4.1. Напишите программу на Python, чтобы нарисовать график рассеяния со случайным распределением по осям X и Y, нанесенный на график друг против друга.

image.png
